$$
n = \text{target size}\\
m = \text{array size}\\
\text{number configs} = \frac{m!}{(m-n)!n!}
$$

```python
import math
from matplotlib.pyplot import *
```

```python
def configs(n,m):
    return math.factorial(m)/(math.factorial(m-n)*math.factorial(n))
```

```python
for x in range(3,10):
    print(x,configs(x,10)**2)
```

```python
x = range(10)
plot(x,[configs(y,10)**2*1e-3 for y in x],'o:')
ylabel('Thousands of Configurations')
xlabel(r'Size of target ($n\times n$)')
grid(True)
ylim(0,70)
savefig('Stiff_Compatible_Configurations.png',transparent=True, bbox_inches='tight', pad_inches=0.1, dpi=400)
```
