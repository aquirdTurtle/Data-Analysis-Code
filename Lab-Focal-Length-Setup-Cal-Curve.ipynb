```python
import numpy as np
from matplotlib.pyplot import *
style.use('seaborn')
```

$$
\frac{1}{x_o} + \frac{1}{x_i} = \frac{1}{f}\\
f = \frac{x_o x_i}{x_o+x_i}
$$

```python
x_t = 1100
x_is = np.linspace(0,1100,1000)
y = np.zeros(len(x_is))
for i, x_i in enumerate(x_is):
    x_o = x_t - x_i
    y[i] = x_o * x_i / (x_o + x_i)
```

```python
figure(figsize=(15,15))
plot(x_is,y, label='Lens-Holder-Face')
plot(x_is,x_is, label='Iris-Image (Look for the Octagon)')
xlabel('Image-Distance(mm)')
ylabel('Focal-Length (mm)')
xlim(0,550)
ylim(0,550)
grid(True, which='both', color='k')
grid(True, which='minor', linestyle=':', color='#090909')
xticks(np.arange(0,550,50));
yticks(np.arange(0,550,50));
minorticks_on()
legend();
title('Determining-Lens-Focal-Length');
```
