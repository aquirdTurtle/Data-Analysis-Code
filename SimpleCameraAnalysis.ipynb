# Simple camera analysis

```python
import Analysis_Python_Files.ExpFile as exp
import matplotlib.pyplot as plt
import numpy as np
exp.setPath('18','October','2020')
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
with exp.ExpFile(16) as file:
    pics = file.get_pics()
    avgPic = file.get_avg_pic()
```

```python
fig, axs = plt.subplots(1,2, figsize=(20,10))
maxX = 5000
heights, binlocs = np.histogram(pics[100:].flatten(), bins=np.arange(maxX), density=True)
im1 = axs[0].imshow(avgPic)
axs[0].set_title('Average Image')
axs[1].semilogy(binlocs[:-1], heights);
axs[1].set_xlim(0, maxX)
axs[1].set_title('Count Hist')
fig.suptitle('EM-Gain x1000');

divider = make_axes_locatable(axs[0])
cax = divider.append_axes('right', size='5%', pad=0.05)
fig.colorbar(im1, cax=cax, orientation='vertical')
```
