# Simple camera analysis

```python
import Analysis_Python_Files.ExpFile as exp
import matplotlib.pyplot as plt
import numpy as np
exp.setPath('22','October','2020')
```

```python
with exp.ExpFile(3) as file:
    pics = file.get_pics()
    avgPic = file.get_avg_pic()
```

```python
len(pics)
```

```python
fig, axs = plt.subplots(1,2, figsize=(20,10))
maxX = 2000
#heights, binlocs = np.histogram(pics[10:].flatten(), bins=np.arange(maxX), density=True)
heights, binlocs = np.histogram(pics[100:].flatten(), bins=np.arange(maxX), density=True)
axs[0].imshow(avgPic)
axs[0].set_title('Average Image')
axs[1].semilogy(binlocs[:-1], heights);
axs[1].set_xlim(0, maxX)
axs[1].set_title('Count Hist')
fig.suptitle('EM-Gain x1000');
```
