Bias Ramp: ldumpBias to (1-tunDumpBias), therefore 0.8 to 0.025 or 0.01, exp ramp

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('expDecay_Niawg.txt') as file:
    line = file.readlines()
```

```python
niawgTimes = np.arange(0,0.4e-3,1/(320e6))[:-1]
niawgInit = (1-0.8)
niawgFin = 0.01
niawgPercents = [float(val) for val in line[0].split(' ')[:-1]]
niawgVals = [niawgInit + per * (niawgFin - niawgInit) for per in niawgPercents]
```

```python
initAg = 0.5
finAg = 0.2
agPercents = []
agTimes = np.arange(0,0.4e-3,1/(500e3))
for time in agTimes:
    agPercents.append((0.5*(1+np.tanh(-4+8*time/(0.4e-3)))))
agVals = [initAg + per * (finAg-initAg) for per in agPercents]
```

```python
totalTimes = [0]
niawgCount = 0
agCount = 0
niawgTime = niawgTimes[0]
agTime = agTimes[0]
totalIntensity = [niawgVals[niawgCount]*agVals[agCount]]
while True:
    # determine which time to increment
    if niawgCount+1 == len(niawgTimes) and agCount+1 == len(agTimes):
        break
    if niawgCount+1 < len(niawgTimes) and niawgTimes[niawgCount+1] < agTimes[agCount+1]:
        # next change is agilent
        niawgCount += 1
        totalTimes.append(niawgTimes[niawgCount])
    else:
        # next change is agilent
        agCount += 1
        totalTimes.append(agTimes[agCount])
    totalIntensity.append(niawgVals[niawgCount]*agVals[agCount])
```

```python
fig, axs = plt.subplots(1,3, figsize=(15,5))
axs[0].plot(niawgTimes*1e3, niawgVals)
axs[0].set_ylim(0,0.2)
axs[0].set_xlabel('Time (ms)')
axs[0].set_ylabel('% Power in Main Tweezers')

axs[1].plot(agTimes*1e3, agVals)
axs[1].set_ylim(0,0.5)
axs[1].set_ylabel("Servo Power Setting (mW)")

axs[2].semilogy(np.array(totalTimes)*1e3,np.array(totalIntensity)/6)
axs[2].set_ylim(0,0.2*0.5/6)
axs[2].set_ylabel("Composite Depth (mK)")

for ax in axs:
    ax.set_xlim(0,0.4)
```

```python
with open('ShallowTweezersRamp.txt', 'w') as file:
    for time, dp in zip(totalTimes, np.array(totalIntensity)/6):
        file.write(str(time) + '\t' + str(dp) + '\n')
```
